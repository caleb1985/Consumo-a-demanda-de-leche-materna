## Combinacion de datasets

```python
band = pd.read_csv('band.csv')
instrum = pd.read_csv('instrument.csv')
band ; instrum
```

```python
# LEFT JOIN

pd.merge(band, instrum, how = 'left', on = 'name')
```

```python
# RIGHT JOIN

band ; instrum
pd.merge(band, instrum, how = 'right', on = 'name')
```

```python
# INNER JOIN

band ; instrum
pd.merge(band, instrum, how = 'inner', on = 'name')
```

```python
# OUTER JOIN

band ; instrum
pd.merge(band, instrum, how = 'outer', on = 'name')
```
